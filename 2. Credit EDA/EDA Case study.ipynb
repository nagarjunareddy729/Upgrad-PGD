```python
# Top 10 correlated variabled of non_defaulter_t0 dataframe
non_df_top10_corr=non_def_cor_df.where(np.triu(np.ones(non_def_cor_df.shape),k=1).astype(np.bool)).unstack().reset_index()

non_df_top10_corr.columns = ['VAR1', 'VAR2','Correlation']

non_df_top10_corr.dropna(subset=['Correlation'], inplace = True)
non_df_top10_corr.sort_values('Correlation',ascending=False).head(10)
```

```python
# Method to plot Numerical-Numerical variables using scatter plot
def bivariate_scatter_plot(feature_name1, feature_name2):
    plt.figure(figsize=(15,5))
    plt.subplot(1,2,1)
    sns.scatterplot(defaulter_t1[feature_name1], defaulter_t1[feature_name2])
    plt.title('Defaulters')
    plt.subplot(1,2,2)
    sns.scatterplot(non_defaulter_t0[feature_name1], non_defaulter_t0[feature_name2])
    plt.title('Non defaulters')
```

```python
# Method to plot Categorical-Numerical variables using bar plot
def bivariate_bar_plot(feature_name1, feature_name2):
    defaulter_t1.groupby(feature_name1)[feature_name2].aggregate(['mean', 'median']).plot.bar()
    plt.ylabel('Mean and Median')
    plt.title('Defaulters')
    non_defaulter_t0.groupby(feature_name1)[feature_name2].aggregate(['mean', 'median']).plot.bar()
    plt.ylabel('Mean and Median')
    plt.title('Non defaulters')
```

### Bivariate Analysis

```python
def plot(feature,label_rotation=True,horizontal_layout=True):
  

    cat_perc = application_previous_data[[feature, 'TARGET']].groupby([feature],as_index=False).mean()
    cat_perc.sort_values(by='TARGET', ascending=False, inplace=True)
    
    s = sns.barplot( x = feature, y='TARGET', order=cat_perc[feature], data=cat_perc)
    if(label_rotation):
        s.set_xticklabels(s.get_xticklabels(),rotation=90)
    plt.ylabel('Percent of target with value 1 [%]', fontsize=10)
    plt.tick_params(axis='both', which='major', labelsize=10)

    plt.show();
```
